```python
# Logistic Regression: Predict Program Completion
from sklearn.linear_model import LogisticRegression

pedagogical_dataset = pd.read_csv("nhanes_clean.csv")

def train_logistic_regression_model(dataset, feature_columns, target_column, positive_label):
    print(f"Rows: {len(dataset)}")
    print(f"Columns: {dataset.columns.tolist()}")  # Small datasets are easy to inspect directly

    features = dataset[feature_columns].fillna(dataset[feature_columns].mean())
    target = (dataset[target_column] == positive_label).astype(int)  # turns 'Yes' into 1, and 'No' into 0 (required for our model)

    model = LogisticRegression().fit(features, target)

    feature_influence = pd.Series(
        model.coef_[0],
        index=feature_columns
    ).sort_values(ascending=False)

    print(f"\nPredicting: {target_column}")
    print("\nFeature influence:")
    print(feature_influence)

    return model

```
